```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
data = np.loadtxt('res.txt', delimiter = ',')
```

```python
mpl.style.use('seaborn')

labels = ['randomized select', 'deterministic select']

fig, ax = plt.subplots(ncols = 2,figsize=(16,7))
ax[0].set_ylabel('runtime (milliseconds)')
ax[0].set_xlabel('input size')
ax[0].plot(data[:,0], data[:,1], marker = "o", color = "orange", label = 'building a tree')
ax[0].legend()
ax[0].set_title('a)')

ax[1].set_ylabel('runtime (nanoseconds)')
ax[1].set_xlabel('input size')
ax[1].plot(data[:,0], data[:,2], marker = "o", color = "green", label = 'rb_insert')
ax[1].plot(data[:,0], data[:,3], marker = "o", color = "magenta", label = 'rb_delete')
ax[1].legend()
ax[1].set_title('b)')

plt.savefig('graph.png', bbox_inches = 'tight', pad_inches = 0)
```
