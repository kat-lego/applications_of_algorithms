```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
data = np.loadtxt('res.txt', delimiter = ',')
```

```python
mpl.style.use('seaborn')

fig, ax = plt.subplots(nrows=1, ncols=1, figsize = (7,7))
ax.set_ylabel('runtime (milliseconds)')
ax.set_xlabel('input size')
ax.plot(data[:,0], data[:,1], marker = "o", color = "orange", label = 'graham scan')
ax.legend()


plt.savefig('1.png', dpi = 400)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize = (14,6))

ax[0].set_ylabel('runtime (milliseconds)')
ax[0].set_xlabel('input size')
ax[0].plot(data[:,0], data[:,2], marker = "o", color = "purple", label = 'jarvis march')
ax[0].plot(data[:,0], data[:,3], marker = "o", color = "red", label = 'jarvis march (runtime/h)')
ax[0].legend()
ax[0].set_title('a)')

ax[1].set_ylabel('h (number of vertices in ConvexHull)')
ax[1].set_xlabel('input size')
ax[1].plot(data[:,0], data[:,4], marker = "o", color = "orange", label = 'graham scan')
ax[1].set_title('b)')

plt.savefig('2.png', dpi = 500)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize = (14,6))

ax[0].set_ylabel('runtime (milliseconds)')
ax[0].set_xlabel('input size')
ax[0].plot(data[:,0], data[:,1], marker = "o", color = "orange", label = 'graham scan')
ax[0].plot(data[:,0], data[:,2], marker = "o", color = "purple", label = 'jarvis march')
ax[0].legend()
ax[0].set_title('a)')


x = data[:,0]
y = data[:,1]/data[:,2]

m,c = np.polyfit(x,y,1)
y_fit = m*x + c
ax[1].set_ylabel('runtime (milliseconds)')
ax[1].set_xlabel('input size')
ax[1].scatter(x, y, marker = "o", color = "blue", label = 'graham scan / jarvis march')
ax[1].plot(x, y_fit, color = "red", label = 'Line of best fit :graham scan / jarvis march')
ax[1].legend()
ax[1].set_title('b)')
plt.savefig('3.png', dpi = 500)
```
