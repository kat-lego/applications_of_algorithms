```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
data = np.loadtxt('res.txt', delimiter = ',')
```

```python
mpl.style.use('seaborn')

labels = ['randomized select', 'deterministic select']

fig, ax = plt.subplots(nrows = 2, ncols = 2,figsize=(16,12))
ax[0][0].set_ylabel('runtime (milliseconds)')
ax[0][0].set_xlabel('input size')
ax[0][0].plot(data[:,0], data[:,1], marker = "o", color = "orange", label = 'gram scan')
ax[0][0].legend()
ax[0][0].set_title('a)')

ax[0][1].set_ylabel('runtime (milliseconds)')
ax[0][1].set_xlabel('input size')
# ax[1].plot(data[:,0], data[:,2], marker = "o", color = "green", label = 'jarvis march')
ax[0][1].plot(data[:,0], data[:,3], marker = "o", color = "magenta", label = 'jarvis march (runtime/h)')
ax[0][1].legend()
ax[0][1].set_title('b)')

ax[1][0].set_ylabel('runtime (milliseconds)')
ax[1][0].set_xlabel('input size')
ax[1][0].plot(data[:,0], data[:,1], marker = "o", color = "orange", label = 'gram scan')
ax[1][0].plot(data[:,0], data[:,2], marker = "o", color = "purple", label = 'jarvis march')
ax[1][0].legend()
ax[1][0].set_title('c)')

x = data[:,0]
y = data[:,1]/data[:,2] 
y_mean = [np.mean(y)]*len(x)
ax[1][1].set_ylabel('runtime (milliseconds)')
ax[1][1].set_xlabel('input size')
ax[1][1].plot(x, y, marker = "o", color = "blue", label = 'gram scan / jarvis march')
ax[1][1].plot(x, y_mean, marker = "o", color = "red", label = 'average :gram scan / jarvis march')
ax[1][1].legend()
ax[1][1].set_title('d)')

plt.savefig('graph.png', bbox_inches = 'tight', pad_inches = 0)
```
