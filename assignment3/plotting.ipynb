```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
data = np.loadtxt('res.txt', delimiter = ',')
```

```python
def plot(data):
    mpl.style.use('seaborn')
    
    labels = ['randomized select', 'deterministic select']

    fig, ax = plt.subplots(ncols = 1,figsize=(12,7))
    
    ax.set_ylabel('runtime (ms)')
    ax.set_xlabel('array input size')
    for j in range(2):
        ax.plot(data[:,1], data[:,2+j],
                 marker = "o",
                 label = labels[j]
                )
    ax.legend()
#     plt.savefig(xlabel+'.png')
```

```python
plot(data)
```
