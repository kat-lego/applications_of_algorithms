```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
data = np.loadtxt('res.txt', delimiter = ',')
```

```python
mpl.style.use('seaborn')

labels = ['randomized select', 'deterministic select']

fig, ax = plt.subplots(ncols = 2,figsize=(16,6))

ax[0].set_ylabel('runtime (ms)')
ax[0].set_xlabel('array input size')
for j in range(2):
    ax[0].plot(data[:,1], data[:,2+j],
             marker = "o",
             label = labels[j]
            )
ax[0].legend()
ax[0].set_title("Runtime vs Array Input size")

x = data[3:,1]
y = data[3:,3]/data[3:,2] 
y_mean = [np.mean(y)]*len(x)

ax[1].set_ylabel('runtime ratio')
ax[1].set_xlabel('array input size')
ax[1].plot(x, y,marker = "o",label = 'randomized:deterministic')

ax[1].plot(x,y_mean, label = 'mean ratio', color = 'red')
ax[1].legend()
ax[1].set_title("The ratio of the Randomized select to the Deterministic select runtimes");
plt.savefig('graphs.png')
```
